#### Project 3:  FizzBuzz Sequence
##### Write a Python program that generates and prints the first 20 numbers in the FizzBuzz sequence.
##### If a number is divisible by 3, print 'Fizz' instead of the number.
##### If a number is divisible by 5, print 'Buzz' instead of the number.
##### If a number is divisible by both 3 and 5, print 'FizzBuzz'.
##### Otherwise, print the number itself.

```python
#  FizzBuzz Sequence
#Write a Python program that generates and prints the first 20

FB_sequence = list(range(1, 21))
print(FB_sequence)



for i in range(1,21):
    
#If a number is divisible by both 3 and 5, print 'FizzBuzz'.    
    if i/3 in FB_sequence and i/5 in FB_sequence:
         print("FizzBuzz")

#If a number is divisible by 3, print 'Fizz' instead of the number.            
    elif i/3 in FB_sequence:
        print("Fizz")

#If a number is divisible by 5, print 'Buzz' instead of the number.
    elif i/5 in FB_sequence:
        print ("Buzz")

#Otherwise, print the number itself.
    else:
        print(i)
```

#### Project 4:  
##### Write a function called 'calculate_grade' that takes a student's score (out of 100 as input and returns their corresponding letter grade according to the following scale: 
##### 90 or above: 'A'
##### 80 to 89: 'B'
##### 70 to 79:'C'
##### 60 to 69:'D'

```python
#Define function: 'calculate_grade' that takes a student's score 
#(out of 100 as input and returns their corresponding letter grade according
#to the following scale

def calculate_grade(stud_score):
   
    if stud_score >= 90:                                                                #90 or above: 'A'
        print("Grade: A")
    
    elif stud_score >= 80:                                                            #80 to 89: 'B'          
        print("Grade: B")
     
    elif stud_score >= 70:                                                            #70 to 79:'C'              
        print("Grade: c")
      
    elif stud_score >= 60:                                                            #60 to 69:'D' 
        print("Grade: D")
       
    else:
        print("Grade: F")                                                              #60 to 69:'D' 
    
```
