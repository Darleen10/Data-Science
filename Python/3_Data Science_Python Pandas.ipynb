#### Groupby and Aggregate functions in Pandas

```python
#import sample dataset
m_flav = pd.read_csv(r"C:\Users\damil\Documents\2024 Skills\Data Science_Python\python-for-data-science-dataset\python-for-data-science-course-main\Flavors.csv")

m_flav.head()
```

```python
m_flavgb = m_flav.groupby("Base Flavor")

m_flavgb.head()
```

```python
## Aggregation:Mean
#Syntax: variable_name = variable_name.groupby(column_name).mean()

m_flavgb = m_flav.groupby("Base Flavor")
m_flavgb.mean(numeric_only = True)
```

```python
## Aggregation:Sum
#Syntax: variable_name = variable_name.groupby(column_name).sum()

m_flav.groupby("Base Flavor").sum(numeric_only = True)

#or

#m_flavgb = m_flav.groupby("Base Flavor")
#m_flavgb.sum(numeric_only = True)
```

```python
## Aggregation 
#for multiple aggregation to one or more columns: AGG FUNCTION

#agg function

m_flav.groupby("Base Flavor").agg({ 
    "Flavor Rating": ["mean", "min", "max", "count"]  
})
```

```python
## Aggregation 
#for multiple aggregation to one or more columns: AGG FUNCTION

#agg function

m_flav.groupby("Base Flavor").agg({ 
    "Flavor Rating": ["mean", "min", "max", "count"],
    "Texture Rating": ["mean", "min", "max", "count"]
})
```

```python
## grouping and Aggregation 
#grouping by more than one column and multiple aggregation to one or more columns: AGG FUNCTION

#agg function

m_flav.groupby(["Base Flavor", "Liked"]).agg({ 
    "Flavor Rating": ["mean", "min", "max", "count"],
    "Texture Rating": ["mean", "min", "max", "count"]
})
```
