```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#import sample file
m_cal = pd.read_csv(r"C:\Users\damil\Documents\2024 Skills\Data Science_Python\python-for-data-science-dataset\python-for-data-science-course-main\calories.csv")
print (m_cal.shape)
m_cal.head(27)
```

```python
#Data Cleaning

##1. dropping null values
m_cal.dropna(inplace = True)
m_cal.info()

##2.convert wrong date format to datetime format
#m_cal["Date"] = pd.to_datetime(m_cal["Date"], errors='coerce')
#m_cal.info()
##4. change index to date column
# Check if 'Date' exists and has no missing values
#if "Date" in m_cal.columns and m_cal["Date"].isna().sum() == 0:
#    m_cal.set_index("Date", inplace=True)
#    print(m_cal.head())
#else:
#    print("Error: The 'Date' column is missing or contains NaN values.")
#
m_cal.head(27)
```

```python
#Creating a line plot

#change figure size
plt.figure(figsize = (20, 7))

#create plot
plt.plot(
    m_cal["Date"], 
    m_cal["Pulse"], 
    linestyle = "--",
    color = "red",
    marker = "o",
    label = "Pulse"
)

# rotate X-axis labels
plt.xticks(rotation = 25)

#add title
plt.title("Daily Pulse Trend")

#labels
plt.xlabel("Date")
plt.ylabel("Pulse")
plt.legend()


plt.show()
```

```python
#Creating a line plot

#change figure size
plt.figure(figsize = (20, 7))

#create plot
plt.plot(
    m_cal["Date"], 
    m_cal["Pulse"], 
    linestyle = "--",
    color = "indigo",
    marker = "v",
    label = "Pulse",
    linewidth = 3
)

# rotate X-axis labels
plt.xticks(rotation = 25)

#add title
plt.title("Daily Pulse Trend")

#labels
plt.xlabel("Date")
plt.ylabel("Pulse")
plt.legend()


plt.show()
```

```python
#Creating a bar plot

#change figure size
plt.figure(figsize = (20, 7))

#create plot
plt.bar(
    m_cal["Date"], 
    m_cal["Maxpulse"], 
    #label = "Maxpulse",
    #    color = "g",
# Create a bar plot with different colors for each bar
color = ['blue', 'indigo']

)

# rotate X-axis labels
plt.xticks(rotation = 45)

#add title
plt.title("Daily Maxpulse Trend")

#labels
plt.xlabel("Date")
plt.ylabel("Maxpulse")
#plt.legend()


plt.show()
```

```python

#Creating a scatter plot

#change figure size
plt.figure(figsize = (14, 7))

#create plot
plt.scatter(
    m_cal["Calories"], 
    m_cal["Pulse"], 
    color = "m",
)

# rotate X-axis labels
plt.xticks(rotation = 25)

#add title
plt.title("Scatter Plot of Calories vs. Pulse")

#labels
plt.xlabel("calories")
plt.ylabel("Pulse")



plt.show()
```
